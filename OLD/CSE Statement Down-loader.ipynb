```python
link0 = "https://www.cse.lk/home/company-info/"
link2 = "/financial"
save0 = "C:/Users/Avin/Downloads/Q1Dump/"
save2 = ".pdf"

periods = ["31.12.2019", "31/12/2019", "31-12-2019", 
           "31st December 2019", "31 December 2019", "31st DECEMBER 2019", 
           "31 DECEMBER 2019", "12/31/2019", "12.31.2019", "12-31-2019"]
```

```python
failed_stocks = []

for stock in stocks_comp:
    link = link0+stock+link2
    driver.get(link)
    
    try:
        myElem = WebDriverWait(driver, 60).until(EC.presence_of_element_located((By.PARTIAL_LINK_TEXT, 'Financial')))
    except TimeoutException:
        print ("Time exceeded")
    
    for period in periods:
        try:
            link_statement = driver.find_element_by_partial_link_text(period)
            break
        except:
            pass
    
    try:
        href = link_statement.get_attribute('href')
        save_path = save0+stock+save2
        file = wget.download(href, save_path)
        time.sleep(1)
        print(save_path)
    except:
        print("Failed", stock)
        failed_stocks.append(stock)
```
